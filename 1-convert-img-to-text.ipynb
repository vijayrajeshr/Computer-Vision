Convert an Colourful Image to GrayScaled and Show the outputs; Also give their matrix forms

```python
import cv2   # for image processing
```

```python

# Load the image (give correct path)
image = cv2.imread("sample-images/tree.jpg")

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Print Original image as matrix
#note; the actual image is an colored such that it will contain 3x3 matrix
print("Original Image Matrix:")
print(image)
```

```python
# Print grayscale image as matrix
print("Grayscale Image Matrix:")
print(gray_image)

# Optional: print shape (rows x columns)
#print("\nMatrix Shape:", gray_image.shape)
```

```python
# Show original image
#cv2.imshow("Original Image", image)

# Show grayscale image
cv2.imshow("Grayscale Image", gray_image)

# Wait until any key is pressed   #cv2.waitKey(1000) 1000---> 1 second
cv2.waitKey(0)
cv2.destroyAllWindows()
```
